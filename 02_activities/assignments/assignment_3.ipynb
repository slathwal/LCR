#### **Question 2:** 
#### Data-visualization

Let's create plots to visualize the relationships between the features (e.g., alcohol, malic acid, ash...).


```python
def plot_feature_pairs(data, feature_names, color_labels=None, title_prefix=''):
    """
    Helper function to create scatter plots for all possible pairs of features.
    
    Parameters:
    - data: DataFrame containing the features to be plotted.
    - feature_names: List of feature names to be used in plotting.
    - color_labels: Optional. Cluster or class labels to color the scatter plots.
    - title_prefix: Optional. Prefix for plot titles to distinguish between different sets of plots.
    """
    # Create a figure for the scatter plots
    plt.figure(figsize=(60, 60))
    
    # Counter for subplot index
    plot_number = 1
    
    # Loop through each pair of features
    for i in range(len(feature_names)):
        for j in range(i + 1, len(feature_names)):
            plt.subplot(len(feature_names)-1, len(feature_names)-1, plot_number)
            
            # Scatter plot colored by labels if provided
            if color_labels is not None:
                plt.scatter(data[feature_names[i]], data[feature_names[j]], 
                            c=color_labels, cmap='viridis', alpha=0.7)
            else:
                plt.scatter(data[feature_names[i]], data[feature_names[j]], alpha=0.7)
            
            plt.xlabel(feature_names[i])
            plt.ylabel(feature_names[j])
            plt.title(f'{title_prefix}{feature_names[i]} vs {feature_names[j]}')
            
            # Increment the plot number
            plot_number += 1

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Show the plot
    plt.show()

# Get feature names
feature_names = wine_df.columns

# Use the helper function to plot scatter plots without coloring by cluster labels
plot_feature_pairs(wine_df, feature_names, title_prefix='Original Data: ')
```

#### **Question 5:** 
#### Bootstrapping 

Implement bootstrapping on the mean of `color intensity`. Generate 10000 bootstrap samples, calculate the mean for each sample, and compute a 90% confidence interval.

```python
# Your answer here...
np.random.seed(123)
color_intensity_samples = []
for i in range(10_000): # run the loop 10,000 times
    sample = wine_df.sample(frac = 1, replace = True) # generate a sample equal in size to the original wine_df with replacement
    sample = sample.assign(replicate = i) # add a new column to the generated sample dataframe with the sample number
    color_intensity_samples.append(sample) # append the generated bootstrapped sample to the list

color_intensity_samples_df = pd.concat(color_intensity_samples) # combine all the bootstrapped samples stored in a list into a dataframe

# calculate the mean color intensity of each bootstrapped sample and store the results in a dataframe
mean_color_intensity_df = color_intensity_samples_df.groupby("replicate")['color_intensity'].mean().reset_index(name = "mean_color_intensity") 

# calculate the 90% confidence interval from the bootstrapped samples distribution
ci_means = mean_color_intensity_df['mean_color_intensity'].quantile([0.05, 0.95])

# calcualte the mean color intensity of the original sample
mean_color_intensity = wine_df['color_intensity'].mean()

# save the upper and lower bounds of confidence intervals into variables
lower_bound = ci_means.iloc[0]
upper_bound = ci_means.iloc[1]

# Display the result
print(f"Mean of Color Intensity: {mean_color_intensity}")
print(f"90% Confidence Interval of Mean Color Intensity: ({lower_bound}, {upper_bound})")

# plot the distribution of means
plt.hist(mean_color_intensity_df["mean_color_intensity"], bins = 20, edgecolor = "black")
```

**Question:**
- Why do we use bootstrapping in this context? What does it help us understand about the mean?

Bootstrapping on mean color intensity of wines in our sample help us understand the variability in the mean of color intensity of wines in the sample data of our wines.

**Question:**
- What is the purpose of calculating the confidence interval from the bootstrap samples? How does it help us interpret the variability of the estimate?

The confidence interval helps us calculate a numeric range for our statistic of interest, which is mean of color intensity in this case. A 90% confidence interval, for example, tells us that we are 90% confident that the true population mean lies within the confidence interval calculated using the bootstrapped samples.

**Question:**

- Reflect on the variability observed in the bootstrapped means and discuss whether the mean of the color intensity appears to be a stable and reliable estimate based on the confidence interval and the spread of the bootstrapped means.

Based on the distribution of bootstrapped means, we see that most means lie around 5 and 5.1, which is the mean of the original samples. The distribution follows a bell shaped curve. Therefore, the mean of color intensity appears to be a stable and reliable estimate from the original wine data sample.
